```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from skopt import BayesSearchCV
from skopt.space import Integer, Real

# Create a small dataset for testing
X_train = np.random.rand(100, 3)  # Example input data with 100 samples and 3 features
y_train = np.random.randint(0, 6, size=(100,))  # Example labels with 6 classes

# Define a function to create and evaluate the neural network
def create_and_evaluate_nn(n_layers, n_nodes, learning_rate):
    model = Sequential()
    model.add(Dense(n_nodes, input_dim=3, activation='relu'))
    
    for _ in range(n_layers - 1):
        model.add(Dense(n_nodes, activation='relu'))
    
    model.add(Dense(6, activation='softmax'))
    
    # Compile the model
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    # Split the small dataset for demonstration
    X_train_split, _, y_train_split, _ = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
    
    # Train the model (in a real scenario, you would use cross-validation)
    model.fit(X_train_split, y_train_split, epochs=5, batch_size=32, verbose=0)
    
    # Evaluate the model and return the negative loss as a score (for minimization)
    loss, _ = model.evaluate(X_train_split, y_train_split, verbose=0)
    
    return -loss

# Define the search space for Bayesian optimization
search_space = {
    'n_layers': Integer(1, 5),
    'n_nodes': Integer(12, 128),
    'learning_rate': Real(1e-6, 1e-2, prior='log-uniform')
}

# Perform Bayesian optimization
bayes_search = BayesSearchCV(
    create_and_evaluate_nn,
    estimator=None,  # Set estimator to None
    scoring='accuracy',  # Specify the scoring metric
    search_spaces=search_space,  
    n_iter=20,  # Number of optimization steps
    cv=3,
    n_jobs=-1,  # Use all available CPU cores
    verbose=1  # Increase verbosity for progress updates
)

# Fit the Bayesian search to your data
bayes_search.fit(X_train, y_train)

# Get the best hyperparameters and best score
best_params = bayes_search.best_params_
best_loss = -bayes_search.best_score_  # Convert back to positive loss

print("Best Hyperparameters:", best_params)
print("Best Validation Loss:", best_loss)
```

```python
# Example for Bayes optimization
from skopt import BayesSearchCV
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X, y = load_digits(n_class=10, return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, test_size=.25, random_state=0)

# log-uniform: understand as search over p = exp(x) by varying x
opt = BayesSearchCV(
    SVC(),
    {
        'C': (1e-6, 1e+6, 'log-uniform'),
        'gamma': (1e-6, 1e+1, 'log-uniform'),
        'degree': (1, 8),  # integer valued parameter
        'kernel': ['linear', 'poly', 'rbf'],  # categorical parameter
    },
    n_iter=32,
    cv=3
)

bayes = opt.fit(X_train, y_train)

print("val. score: %s" % opt.best_score_)
print("test score: %s" % opt.score(X_test, y_test))
best_params = bayes.best_params_
best_loss = -bayes.best_score_  # Convert back to positive loss

print("Best Hyperparameters:", best_params)
print("Best Validation Loss:", best_loss)
```
